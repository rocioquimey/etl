```python
import io
import sys
import pandas as pd
import requests
import xml.etree. ElementTree as ET
urlvideo='https://www.youtube.com/feeds/videos.xml?channel_id=UCZBR7jzRamjhK7FCV9bg-7w'   #LOOOP EN VEZ DE CADA
r=requests.get(urlvideo, headers={ "User-Agent": "Chrome/50.0.2661.94" })
data=r.content
root= ET. fromstring (data.decode())
lista= root.findall('.//')

# Creamos dos objetos StringIO para capturar la salida de los dos prints
captura_titulo = io.StringIO()
captura_visualizaciones = io.StringIO()
captura_url = io.StringIO()
captura_likes=io.StringIO()
captura_desc=io.StringIO()
# Redirigimos la salida estándar al objeto StringIO de cada captura
sys.stdout = captura_titulo
# Ejecutamos el bucle para capturar los títulos
for item in lista:
    titulo = item.find('{http://search.yahoo.com/mrss/}title')
    if titulo is not None:
        print(titulo.text)

sys.stdout = captura_visualizaciones
# Ejecutamos el bucle para capturar las visualizaciones
for item in lista:
    visto = item.get('views')
    if visto is not None:
        print(visto)


sys.stdout = captura_url
for item in lista:
    url=item.get('href')
    if url is None:
       pass
    else:
      print(url)


sys.stdout =captura_likes

for item in lista:
   like=item.get('count')
   if like is None:
    pass
   else:
    print(like)


# Restauramos la salida estándar
sys.stdout = sys.__stdout__

# Obtenemos el contenido capturado en los objetos StringIO y lo guardamos en variables
resultado_titulo = captura_titulo.getvalue()
resultado_visualizaciones = captura_visualizaciones.getvalue()
resultado_url = captura_url.getvalue()
resultado_likes=captura_likes.getvalue()

# Dividimos las cadenas capturadas en líneas
# Dividimos las cadenas capturadas en líneas
lineas_titulo = resultado_titulo.split('\n')
lineas_visualizaciones = resultado_visualizaciones.split('\n')
lineas_url = resultado_url.split('\n')
lineas_url = lineas_url[2:]  # Excluir las primeras dos líneas
lineas_likes = resultado_likes.split('\n')

# Creamos un DataFrame de pandas con las líneas capturadas
df_20LA = pd.DataFrame({'titulo': lineas_titulo, 'visualizaciones': lineas_visualizaciones, 'likes': lineas_likes, 'url': lineas_url,'Channel':'20th Century Studios LA'})

# Imprimimos el DataFrame
df_20LA
```
