#Union schedule con Scrapping YT

```python
import io
import sys
import pandas as pd
import requests
import xml.etree.ElementTree as ET
import pandas as pd
import openpyxl
!pip install schedule
import schedule
import time
from datetime import datetime
import pytz

# Definici√≥n de la funci√≥n de scraping web
def hacer_scraping():
    # Lista de identificadores de canal    #ME FALTA MARVEL- SUMAR DOBLADO Y DLA O LO QUE SEA
    canales = ['UCTFbGAgp5L6IgUTQDW93zlQ', 'UCZBR7jzRamjhK7FCV9bg-7w','UC8huWhDjI8UJK4bqo8BJ9WQ','UCP6nJ-Elnfnpjv6HQRsv1Cw']

    # URL base
    url_base = 'https://www.youtube.com/feeds/videos.xml?channel_id='

    # Lista para almacenar las respuestas de las peticiones
    respuestas = []

    # Iterar sobre cada identificador de canal
    for canal in canales:
        # Construir la URL para el canal actual
        url_canal = url_base + canal
        r = requests.get(url_canal, headers={ "User-Agent": "Chrome/50.0.2661.94" })
        data = r.content
        root = ET.fromstring(data.decode())
        lista = root.findall('.//')

        # Lista para almacenar los datos de cada iteraci√≥n
        data_titulo = []
        data_visualizaciones = []
        data_url = []
        data_likes = []

        for item in lista:
            titulo = item.find('{http://search.yahoo.com/mrss/}title')
            visto = item.get('views')
            url = item.get('href')
            like = item.get('count')

            if titulo is not None:
                data_titulo.append(titulo.text)
            if visto is not None:
                data_visualizaciones.append(visto)
            if url is not None:
                data_url.append(url)
            if like is not None:
                data_likes.append(like)

        if canal == 'UCTFbGAgp5L6IgUTQDW93zlQ':
            canal_para_df = '20th Century Studios BR'
        elif canal == 'UCZBR7jzRamjhK7FCV9bg-7w':
            canal_para_df = '20th Century Studios LA'
        elif canal =='UC8huWhDjI8UJK4bqo8BJ9WQ':
            canal_para_df='Disney Studios LA'
        elif canal =='UCP6nJ-Elnfnpjv6HQRsv1Cw':
            canal_para_df='Disney Studios BR'

        timestamp = datetime.now(pytz.timezone('America/Argentina/Buenos_Aires')).strftime('%Y-%m-%d %H:%M:%S')

        # Crear un DataFrame de pandas con los datos de esta iteraci√≥n
        df_temp = pd.DataFrame({'titulo': data_titulo, 'visualizaciones': data_visualizaciones, 'likes': data_likes, 'url': data_url[2:], 'Channel': canal_para_df,'Execution':timestamp})

        # Agregar el DataFrame temporal a la lista de respuestas
        respuestas.append(df_temp)

    # Concatenar todos los DataFrames en uno solo
    df_final = pd.concat(respuestas, ignore_index=True)

    # Imprimir el DataFrame final
    print(df_final)
    df_final.to_csv('datos_youtube.csv', index=False)

schedule_time = str(input("Ingrese la hora programada (formato HH:MM): "))
fecha_datetime = datetime.strptime(schedule_time, '%H:%M')
fecha_date = fecha_datetime.time()

# Asignar la funci√≥n de scraping web a la variable mi_funcion
mi_funcion = hacer_scraping

# Programar la ejecuci√≥n de la funci√≥n
schedule.every().day.at(schedule_time).do(mi_funcion)

# Ejecutar el c√≥digo indefinidamente
while True:
    # Obtener la fecha y hora actual
    now = datetime.now(pytz.timezone('America/Argentina/Buenos_Aires')).replace(tzinfo=None)

    # Calcular cu√°nto tiempo falta hasta la fecha programada
    tiempo_restante = datetime(now.year, now.month, now.day, fecha_date.hour, fecha_date.minute) - now

    tiempo_restante_segundos = tiempo_restante.total_seconds()
    tiempo_restante_minutos = tiempo_restante.total_seconds()/60
    tiempo_restante_horas = tiempo_restante.total_seconds()/3600

    if tiempo_restante.total_seconds() <= 0:
        mi_funcion()
        break  # Salir del bucle while
    else:
        # Esperar hasta la pr√≥xima ejecuci√≥n programada
        print(f"Esperando hasta las {schedule_time}. El tiempo restante es de {tiempo_restante_horas}      {tiempo_restante_minutos}minutos. segundos ...")
        time.sleep(tiempo_restante_segundos)
        schedule.run_pending()  # Ejecutar las tareas programadas
```
