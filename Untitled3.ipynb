```python

import schedule
import time
from datetime import datetime
import pytz # pytz.all_timezones

schedule_time=str(input( ))
fecha_datetime = datetime.strptime(schedule_time, '%H:%M')
fecha_date = fecha_datetime.time()
def mi_funcion():
    print('¡Es hora! Esta es la tarea programada para el momento ingresado:',schedule_time)

# Programar la ejecución de la función todos los días a las-- fecha programada
schedule.every().day.at(schedule_time).do(mi_funcion)

# Obtener la fecha y hora actual
# now = datetime.now().replace(tzinfo=None)

now = datetime.now(pytz.timezone('Etc/GMT+3')).replace(tzinfo=None)

# Calcular cuánto tiempo falta hasta la ---fecha programada
tiempo_restante = datetime(now.year, now.month, now.day, fecha_date.hour, fecha_date.minute) - now

tiempo_restante_minutos = tiempo_restante.total_seconds()/60
tiempo_restante_segundos = tiempo_restante.total_seconds()


if tiempo_restante.total_seconds() <= 0:
    mi_funcion()
else:
# Esperar hasta la próxima ejecución programada
    print(f"Esperando hasta las {schedule_time} ({tiempo_restante_minutos} minutos.segundos restantes)...")
    time.sleep(tiempo_restante_minutos)

# Ejecutar el código indefinidamente
while True:
    schedule.run_pending()
    time.sleep(1)
```
